#**Exploratory Data Analysis (EDA) for a Sample Dataset**



```python
# We can use the pandas library in python to read in the csv file.
import pandas as pd
#for numerical computaions we can use numpy library
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# This creates a pandas dataframe and assigns it to the titanic variable.
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url,index_col=False)
# Print the first 5 rows of the dataframe.
titanic
```

**VIOLIN PLOT**

Violin Plot (Variable_1 vs. Variable_2 by Category):

The violin plot displays the distribution of Variable_2 categorized by Variable_1.
It provides insights into the distribution of a numerical variable (Variable_2) across different categories or groups (Variable_1).


```python
sb.violinplot(x="Survived", y="Age", hue='Sex', data=titanic)
plt.show()
```

**Inference:**
Maximum men who did not survive lie between the age group of 20-40.

Maximum females who survived were between the age group of 20-40.

**CORRELATION HEATMAP**

Correlation Heatmap:

A heatmap visualizes the correlation matrix of numerical variables in the dataset.
It helps identify relationships between features, with values ranging from -1 (negative correlation) to 1 (positive correlation).
The annotations indicate the strength of correlations between variables.

```python
plt.subplots(figsize=(5, 5))
sb.heatmap(titanic.corr(), cmap = "YlGnBu", annot=True, fmt=".2f")
plt.show()
```

**Inference:**
As the target variable is Survived , the feature with most correlation is FARE

**JOINT PLOT**

Joint Plot (Variable_3 vs. Variable_4):

The joint plot displays a scatter plot of Variable_3 against Variable_4.
It provides a visual representation of the relationship between two numerical variables, highlighting any trends or patterns.

```python
sb.jointplot(x='Age',y='Fare',data=titanic)
plt.show()
```

**INFERENCE:** People of all age groups bought ticket worth minimum fare(10-20) while only some of them bought the most expensive tickets (500+)
The people between age group (20-25) bought the most number of tickets with distinct fare tickets.

**BOX PLOT**

Box Plot (Variable_1 vs. Variable_2):

The box plot illustrates the distribution of Variable_2 based on different values of Variable_1.
It helps identify the central tendency and spread of Variable_2 for each category of Variable_1.

```python
plt.figure(figsize=(5,5))
sb.boxplot(x="Survived", y="Age", hue='Sex', data=titanic)
plt.show()
```

**Inference:** It shows the the area where the accumulation is the most also the distribution inside it.
The males who survived are in the range of 20-40 in which one third lie between 20-30 while the remaining two third lie between 30-40.
